```python
frame = cv2.imread("ironman.jfif")
# cv2 uses bgr instead of rgb therefore opposite collor code
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x,y,w,h =  face
```

```python
faceimg = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(faceimg)
```

```python
out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),4)
```

```python
plt.imshow(out)
```
