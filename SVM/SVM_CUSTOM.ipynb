```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X,y = make_classification(n_classes = 2, n_samples=400, n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,random_state=3)
```

```python
# in SVm we have y = -1 0r 1
# but here we have y==0 or 1

# convert all y=0 to -1
y[y==0] = -1
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
class SVM:
    def __init__(self,C=1.0):
        self.C = C # penalty 
        self.W = 0
        self.b = 0
    
    def hingeloss(self,W,b,X,y):
        loss =0.0
        
        loss += 0.5*np.dot(W,W.T)
        
        for i in range(X.shape[0]):
            ti = y[i] * (np.dot(W,X[i].T)+b)
            loss += self.C * max(0,1-ti)
        
        return loss[0][0] # since we have used np.dot, losss will a matrix like [[loss]] and therefore we only want a scaler value like loss[0][0]
    
    def fit(self,X,y,batch_size =100,lr= 0.01, maxItr=300):
        no_of_features = X.shape[1]
        no_of_samples = X.shape[0]
        n = lr
        c = self.C
        
        #intialise the model parameters
        W = np.zeros((1,no_of_features))
        bias = 0
        
        losses = []
        
        # training loop
        
        for i in range(maxItr):
            l = self.hingeloss(W,bias,X,y)
            losses.append(l)
            
            # for bacth size, get random points
            
            ids = np.arange(no_of_samples)
            np.random.shuffle(ids)
            
            # gradient descend batch wise with random shuffling
            
            for batch_start in range(0,no_of_samples,batch_size):
                # assume gradient is 0 for the batch at first
                gradw = 0
                gradb = 0
                
                # iterate over all the examples in the minbatch
                for j in range(batch_start,batch_start+batch_size):
                    if j< no_of_samples:
                        i = ids[j]
                        ti = y[i] * (np.dot(W,X[i].T)+bias)
                        
                        if ti>=1:
                            gradw +=0
                            gradb+=0
                        else:
                            gradw += c* y[i] * X[i]
                            gradb += c* y[i]
                            
                W = W- n*W + n*gradw
                bias = bias + n*gradb
        self.W = W
        self.b = bias
        return W,bias,losses
    
                
```

```python
model = SVM(C=10)
```

```python
W,b,losses = model.fit(X,y,maxItr=100)
```

```python
plt.plot(losses)
```

```python
def plot_hyperplane(w1,w2,b):
    plt.figure(figsize=(12,12))
    x1 = np.linspace(-2,4,10)
    x2 = -(w1*x1 + b)/w2
    xp = -(w1*x1 +b+1)/w2
    xn = -(w1*x1 +b-1)/w2
    
    plt.plot(x1,x2, label="WX+B=0")
    plt.plot(x1,xp, label="WX+B=1")
    plt.plot(x1,xn, label="WX+B=-1")
    plt.legend()
    
    plt.scatter(X[:,0],X[:,1],c=y)
    plt.show()
```

```python
plot_hyperplane(W[0,0],W[0,1],b)
```

```python
model_1 = SVM(C=80) # margin has decreased as we increase the penalty
W_1,b_1,losses_1 = model_1.fit(X,y,maxItr=100)
plot_hyperplane(W_1[0,0],W_1[0,1],b_1)
plt.plot(losses_1)
```
